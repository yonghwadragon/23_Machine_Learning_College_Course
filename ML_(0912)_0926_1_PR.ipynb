```python
#creditset.csv 파일을 읽으셔서 credit 데이터프레임을 만들기
#데이터프레임에 결측치 있는지 확인
#결축치가 있으면 각 변수의 평균, 또는 최빈값으로 채워넣으세요.
#loan, LTL, income 변수에 대해서 MinMaxScaling을 적용
#train, test로 파티셔닝
#default10yr변수의 1은 "default",0은 "normar"로 값을 치환

#imputing -> fillna(평균이나 중앙값이 들어감)
#         -> fillna(mode)
```

```python
# 판다스(Pandas) 라이브러리를 불러옵니다.
import pandas as pd

# 'creditset.csv'라는 파일에서 데이터를 읽어옵니다.
credit = pd.read_csv("creditset.csv")

# 결측값이 있는 열들을 확인합니다.
# 결측값이란, 데이터가 빠져있는 것을 의미합니다.
# 여기서는 'income' 열에 결측값이 있는지 확인합니다.
credit.isnull().mean()

# 'income' 열에 결측값이 있다면, 그 값을 어떤 값으로 채울 것인지 정합니다.
# 예를 들어, 결측값을 0으로 채우려면 아래와 같이 합니다.
# credit["income"].fillna(0, inplace=True)

# 사이킷런(Scikit-Learn) 라이브러리에서 MinMaxScaler를 불러옵니다.
from sklearn.preprocessing import MinMaxScaler

# 'loan', 'age', 'income', 'LTI' 열을 MinMaxScaler를 사용하여 스케일링(변환)합니다.
scaled = MinMaxScaler().fit_transform(credit[["loan", "age", "income", "LTI"]])

# 스케일링한 결과를 데이터프레임으로 만듭니다.
# 이 데이터프레임은 'loan', 'age', 'income', 'LTI' 열을 가지고 있습니다.
df1 = pd.DataFrame(scaled, columns=["loan", "age", "income", "LTI"])

# 'clientid'와 'default10yr' 열을 따로 저장합니다.
df2 = credit[["clientid", "default10yr"]]

# 'df1'과 'df2' 데이터프레임을 이어 붙여서 하나의 데이터프레임으로 만듭니다.
credit = pd.concat([df1, df2], axis=1)

# 사이킷런의 train_test_split을 사용하여 데이터를 훈련 세트(train)와 테스트 세트(test)로 나눕니다.
# 여기서는 전체 데이터의 30%를 테스트 세트로 사용하며, 'default10yr' 열을 기준으로 분리합니다.
from sklearn.model_selection import train_test_split
train, test = train_test_split(credit, test_size=0.3, stratify=credit.default10yr)

# 'default10yr' 열의 값을 1을 'default'로, 0을 'normal'로 바꿉니다.
credit.default10yr = credit.default10yr.map({1: "default", 0: "normal"})
credit

scaled
```
